```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

haltes = pd.read_excel('haltes.xlsx')

print(len(haltes))
haltes.head()
```

```python
haltes = haltes[haltes.quaycode.str[0:2]=='NL'] # Haltes!A:A;2)="NL"
haltes['geheel_toegankelijk'] = (haltes.quaydisabledaccessible=='Y') & (haltes.quayvisuallyaccessible=='Y')
```

```python

bushaltes = haltes[haltes.transportmode=='bus'] # Haltes!$B:$B=Percentages!$B$1
bushaltes = bushaltes[bushaltes.quaytype=='regular'] # Haltes!$C:$C="regular"
bushaltes = bushaltes[bushaltes.quaystatus=='available'] # Haltes!$D:$D="available"
print('Totaal aantal bushaltes (NL,bus,regular,available): {}'.format(len(bushaltes)))


# Haltes!$E:$E=Percentages!$A4
```

```python
def calculate_table(df,column):
    table = df[['concessionprovidercode',column]].groupby(['concessionprovidercode',column]).size().unstack(fill_value=0)
    ncols=table.columns
    if 'Y' in ncols:
        cols = ['Y','N']
    else:
        cols = [True,False]
    table['sum'] = df.quaycode.groupby(df['concessionprovidercode']).count()
    table['percentage'] = 100*(table[cols[0]] / table['sum'])
    row = []
    for c in ncols:
        row.append(table[c].sum())
    row.append(table['sum'].sum())
    row.append(100*(table[cols[0]].sum() / table['sum'].sum()))
    row = pd.DataFrame([row],columns=table.columns,index=['totaal'])
    table=table.append(row)
    table['percentage'] = table['percentage'].round(0)
    return table
    

calculate_table(bushaltes,'quaydisabledaccessible')
```

```python
calculate_table(bushaltes,'quayvisuallyaccessible')
```

```python
calculate_table(bushaltes,'geheel_toegankelijk')
```

```python
tramhaltes = haltes[haltes.transportmode=='tram'] # Haltes!$B:$B=Percentages!$B$1
tramhaltes = tramhaltes[tramhaltes.quaytype=='regular'] # Haltes!$C:$C="regular"
tramhaltes = tramhaltes[tramhaltes.quaystatus=='available'] # Haltes!$D:$D="available"
print('Totaal aantal tramhaltes (NL,tram,regular,available): {}'.format(len(tramhaltes)))
```

```python
from IPython.display import display
display(calculate_table(tramhaltes,'quaydisabledaccessible'))
display(calculate_table(tramhaltes,'quayvisuallyaccessible'))
display(calculate_table(tramhaltes,'geheel_toegankelijk'))
```
